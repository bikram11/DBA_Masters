```python
import pandas as pd
import os
df = pd.DataFrame()

data_path = '/media/bikram/Samsung_T5/DBA/filtered_dataset/'

data_files=os.listdir(data_path)

data_files.sort()


for items in data_files:
    temp_df = pd.read_csv(data_path+items)
    temp_df = temp_df.drop([ 'Unnamed: 0'], axis=1)

    df=pd.concat([df,temp_df], ignore_index=True)

```
